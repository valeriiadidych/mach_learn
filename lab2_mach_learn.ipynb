```python
# few rows out of table

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display

train_df = pd.read_csv("application_train.csv")
print(train_df.shape)
pd.set_option('display.max_columns', 122)
display(train_df)
```

```python
# another number of rows, but in random order

display(train_df.sample(10))
```

```python
# table showing number of missing values

for column_name in train_df:
    if any([substring in column_name for substring in ("TYPE", "FLAG", "CODE_GENDER")]):
        train_df[column_name] = train_df[column_name].astype("category")

missing_values_info = [[column_name,train_df[column_name].dtype, value] for column_name, value in train_df.isnull().sum().iteritems()]
display(pd.DataFrame(missing_values_info, columns=["feature", "data type", "missing"]).sort_values(ascending=False, by="missing"))
```

```python
# statictic by columns

display(train_df.describe(include="all"))
```

```python
# target classes histogram

sns.histplot(train_df["TARGET"], kde=False)
```

```python
# histogram of categorical features

categorical = train_df.select_dtypes(include=["category"]).columns

plt.rcParams["figure.figsize"] = (5, 3)
for i, col in enumerate(categorical):
    if col == "target":
        continue
    sns.histplot(x=train_df[col], kde=False)
    plt.show()
```

```python
# distributions of all the numerical features

train_df.select_dtypes(exclude=["category", "object"]).hist(figsize=(30, 30))
```

```python
# checking data for outliers

numerical = train_df.select_dtypes(exclude=["category", "object"]).columns
plt.rcParams["figure.figsize"] = (5, 3)
for i, col in enumerate(numerical):
    if col == "target":
        continue
    sns.boxplot(x=train_df[col])
    plt.show()
```

```python
# correlation matrix

correlation_matrix = train_df.corr()
mask = np.array(correlation_matrix)
mask[np.tril_indices_from(mask)] = False
fig, ax = plt.subplots()
fig.set_size_inches(25, 25)
sns.heatmap(correlation_matrix, cmap="Blues", mask=mask, annot=True)
```

```python
# standardizing numerical features, encoding category features

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

def transform_table(df):
    numerical_features = df.select_dtypes(exclude=["category", "object"]).columns
    category_features = df.select_dtypes(exclude=["int64", "float64"]).columns

    transformer = ColumnTransformer(
     transformers = [
         ('numerical', StandardScaler(), numerical_features),
         ('categorical', OneHotEncoder(), category_features)
     ],
)
    return pd.DataFrame(transformer.fit_transform(df))


train_x = transform_table(train_df.drop("TARGET", axis=1))
train_y = train_df["TARGET"]
display(train_x.head())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import HistGradientBoostingClassifier

model = HistGradientBoostingClassifier()
scores = cross_val_score(model, train_x, train_y, scoring="roc_auc")
print(scores)
```

```python
# checking model on train set

from sklearn.metrics import accuracy_score, roc_auc_score

predicted_y = model.predict(train_x)

print(accuracy_score(train_y, predicted_y))
print(roc_auc_score(train_y, predicted_y))
```
